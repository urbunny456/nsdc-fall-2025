#### Combine datasets so there are three total: Common Data Set, UCLA Grade Distributions, and UC Berkeley Grade Distributions

There are three main methods to combine datasets: concat(), merge(), and join(). Feel free to read documentation on them, but I will give a brief summary of what each of them do and how they differ.

concat(): This combines along columns (you add columns that don't exist previously) or rows (add rows that don't exist previously). This is kind of similar to the SQL join method, where you can choose inner join (only get columns/indices that are shared) or outer join (return everything)

merge(): This is even closer to the SQL join method, where there's left, right, outer, inner, cross join. This is primarily used for joining through columns.

join(): Also similar to SQL join method, with left, right, outer, etc. This is mainly for combining datasets through their indices. 

### Common Data Set

Since we are ADDING NEW ROWS based on existing columns and not adding new columns based on existing rows, we can use concat

```python
common_data_set = pd.concat([ucla_cds, berk_cds, harvard_cds], ignore_index=True)
common_data_set.head(20)

common_data_set.to_csv('common_data_set.csv', encoding='utf-8', index=False)
```
